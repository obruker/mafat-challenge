```python
# Fix the results

segments = ['sunroof','luggage_carrier','open_cargo_area','enclosed_cab','spare_wheel','wrecked','flatbed','ladder','enclosed_box','soft_shell_box','harnessed_to_a_cart','ac_vents']

df_answer = pd.read_csv('results/answer.csv')
for segment in segments:
    results = pd.read_csv('results/results_{}.csv'.format(segment))
    sorted_by_prediction = results.sort_values(by='1', ascending=False)
    tags = sorted_by_prediction['tag_id'].tolist()
    df_answer[segment] = tags

df_answer.drop(['0', '1'], axis=1, inplace=True)
df_answer.to_csv('results/answer_fixed.csv',index=False)
df_answer.head()
```
